```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open('11.webp').convert('L')
img_array = np.array(img)
# we are adding some constant to avoid log of zero error
log_img = np.log1p(img_array)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(log_img, cmap='gray')
plt.title('Log Domain Image')
plt.axis('off')
plt.show()
```

```python
def convolve2d(image, kernel):
    k_h, k_w = kernel.shape
    i_h, i_w = image.shape
    pad_h, pad_w = k_h // 2, k_w // 2
    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')
    output_image = np.zeros_like(image)

    for i in range(i_h):
        for j in range(i_w):
            roi = padded_image[i:i + k_h, j:j + k_w]
            # Perform  multiplication and sum for each elements
            output_image[i, j] = np.sum(roi * kernel)

    return output_image

kernel_size = 15
# overall we are passing 15 x 15 kernel(filter) to every pixel
low_pass_kernel = np.ones((kernel_size, kernel_size)) / (kernel_size**2)

low_freq_img = convolve2d(log_img, low_pass_kernel)

high_freq_img = log_img - low_freq_img
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(log_img, cmap='gray')
plt.title('Log Domain Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(low_freq_img, cmap='gray')
plt.title('Low Frequency Component (Illumination)')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(high_freq_img, cmap='gray')
plt.title('High Frequency Component (Reflectance)')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
color_img = Image.open('22.webp').convert('RGB')
color_img_array = np.array(color_img)
# we are seperating rgb values
r_channel = color_img_array[:, :, 0]
g_channel = color_img_array[:, :, 1]
b_channel = color_img_array[:, :, 2]
log_r = np.log1p(r_channel)
log_g = np.log1p(g_channel)
log_b = np.log1p(b_channel)

# we are applying low pass filters for these channels of rgb
low_freq_r = convolve2d(log_r, low_pass_kernel)
low_freq_g = convolve2d(log_g, low_pass_kernel)
low_freq_b = convolve2d(log_b, low_pass_kernel)

# high frequency domains
high_freq_r = log_r - low_freq_r
high_freq_g = log_g - low_freq_g
high_freq_b = log_b - low_freq_b

# converting those log values back to (x,y) format
corrected_r = np.expm1(high_freq_r)
corrected_g = np.expm1(high_freq_g)
corrected_b = np.expm1(high_freq_b)
corrected_color_img_array = np.stack([corrected_r, corrected_g, corrected_b], axis=2)

# Normalizing and scaling values before converting to uint8
min_val = np.min(corrected_color_img_array)
max_val = np.max(corrected_color_img_array)
scaled_color_img_array = (corrected_color_img_array - min_val) / (max_val - min_val) * 255
scaled_color_img_array = np.clip(scaled_color_img_array, 0, 255).astype(np.uint8)
plt.figure(figsize=(10, 5)) # here we are Displaying the original and corrected color images
plt.subplot(1, 2, 1)
plt.imshow(color_img_array)
plt.title('Original Color Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(scaled_color_img_array)
plt.title('Corrected Color Image (Reflectance)')
plt.axis('off')
plt.show()
```
